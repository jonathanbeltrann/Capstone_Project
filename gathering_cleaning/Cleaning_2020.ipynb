## Cleaning Data: 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from datetime import datetime
```

```python
taxi_2020 = pd.read_csv('./data/taxi_2020.csv')
```

```python
taxi_2020.dtypes
```

```python
taxi_2020 = taxi_2020.drop(columns= 'Unnamed: 0')
```

```python
taxi_2020 = taxi_2020.dropna()
```

```python
## Changing pickup and drop off to datetime as well as the name

taxi_2020['tpep_pickup_datetime'] = (pd.to_datetime(taxi_2020['tpep_pickup_datetime']))
taxi_2020['tpep_dropoff_datetime'] = (pd.to_datetime(taxi_2020['tpep_dropoff_datetime']))
```

```python
## Renaming pickup/ dropoff datetime columns
taxi_2020 = taxi_2020.rename(columns={'tpep_pickup_datetime': 'pickup_datetime', 'tpep_dropoff_datetime': 'dropoff_datetime'})
```

```python
## Extracting Day of the week from pickup/dropoff into columns
taxi_2020['pickup_day_of_week'] = taxi_2020['pickup_datetime'].dt.day_name()
taxi_2020['dropoff_day_of_week'] = taxi_2020['dropoff_datetime'].dt.day_name()
```

```python
## Extracting hour of the day from pickup/dropoff into columns
taxi_2020['pickup_hour']= taxi_2020['pickup_datetime'].dt.hour
taxi_2020['dropoff_hour']= taxi_2020['dropoff_datetime'].dt.hour
```

```python
taxi_2020.dtypes
```

```python
taxi_2020 = taxi_2020.loc[taxi_2020['pickup_datetime'].dt.year == 2020,:]
```

```python
taxi_2020.shape
```

```python
## Checking to see all dates belong to 2020
taxi_2020['pickup_datetime'].dt.year.value_counts()
```

```python
##Function binning hours of the day for morning, afternoon and late night from pickup/dropoff hour of the day column
def time_of_day(x):
    if x in range(6,12):
        return 'Morning'
    elif x in range(12,16):
        return 'Afternoon'
    elif x in range(16,22):
        return 'Evening'
    else:
        return 'Late night'
```

```python
taxi_2020['pickup_timeofday']=taxi_2020['pickup_hour'].apply(time_of_day)
```

```python
taxi_2020['dropoff_timeofday']=taxi_2020['dropoff_hour'].apply(time_of_day)
```

```python
### trip time

taxi_2020['trip_time'] = taxi_2020['dropoff_datetime'] - taxi_2020['pickup_datetime']
```
